```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
pd.set_option('future.no_silent_downcasting', True)
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import seaborn as sns
```

```python
df= pd.read_csv("music_genre.csv")
df.dropna(inplace=True)
df.info()
```

```python
df.drop(columns=['instance_id','artist_name','track_name','obtained_date'], inplace=True)
for i in df.columns:
    if((df[i]=='?').any()):
        print(i)

(df.tempo== '?').value_counts()
df.replace('?',np.nan,inplace=True)
df.dropna(inplace=True)
df
```

```python
# Standardization 
df['duration_ms']= df['duration_ms']/(3600)
scaler= StandardScaler()
for i in df.columns:
    if(i== 'key' or i== 'mode' or i== 'music_genre'):
        continue
    else:
        df[i]= scaler.fit_transform(df[[i]])

df
```

```python
# One hot encoding the features
df.replace({'Major':0, 'Minor':1}, inplace=True)
key_list=df['key'].unique()
encode= OneHotEncoder(drop='first', sparse_output=False)
encode_df= encode.fit_transform(df[['key']])
encode_df
for i in range(0, len(key_list)-1):
    df[key_list[i]]= encode_df[:,i]

df
```

```python
# One hot encoding the labels
df.drop(columns='key', inplace=True)
df['music_genre_encoded']= LabelEncoder().fit_transform(df['music_genre'])
df.drop(columns='music_genre', inplace=True)
df
```

```python
X_arr, Y_arr= df.drop(columns='music_genre_encoded'), df['music_genre_encoded']
X_cols= X_arr.columns
for i in range(0,len(X_arr.columns)-11):
    sns.histplot(x= X_arr[X_cols[i]], data=X_arr)
    plt.show()
```

```python
sns.histplot(x= Y_arr)
Y_arr.value_counts()
```

```python
plt.figure(figsize=[15,7])
sns.heatmap(X_arr.corr().abs(),annot=True )
df.drop(columns=['acousticness','loudness','instrumentalness'], inplace=True)
```

```python
df.to_csv("Modified_data.csv", index=False)
```
