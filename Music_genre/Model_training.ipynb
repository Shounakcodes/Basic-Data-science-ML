```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
import pickle
```

```python
df= pd.read_csv('Modified_data.csv')
df
```

```python
X_arr, Y_arr= df.drop(columns='music_genre_encoded'), df['music_genre_encoded']
X_train, X_test, Y_train, Y_test= train_test_split(X_arr,Y_arr, test_size=0.2)

param1= {'max_iter':[100,200,500,1000],'C':[0.01,0.1,0.5,1,5,10,100]}
param2= {'criterion':['gini','entropy','log_loss'],'max_depth':[1,3,5,10,20],'ccp_alpha':[0,0.5,1,5,10]}
param3= { 'n_estimators':[1,3,5,10,20]}
model1= LogisticRegression()
model2= DecisionTreeClassifier()
model3= RandomForestClassifier()
model4= GaussianNB()

cv1= GridSearchCV(model1, param_grid=param1, scoring='accuracy')
cv2= GridSearchCV(model2, param_grid=param2,scoring='accuracy')
cv3= GridSearchCV(model3, param_grid=param3,scoring='accuracy')

```

```python
cv1.fit(X_train, Y_train)
pred1=cv1.predict(X_test)
print(accuracy_score(Y_test, pred1))
print(classification_report(Y_test,pred1))
```

```python
cv2.fit(X_train, Y_train)
pred2=cv2.predict(X_test)
print(accuracy_score(Y_test, pred2))
print(classification_report(Y_test,pred2))
```

```python
cv3.fit(X_train, Y_train)
pred3=cv3.predict(X_test)
print(accuracy_score(Y_test, pred3))
print(classification_report(Y_test,pred3))
```

```python
def k_folds(X,Y):
    kf= KFold(n_splits=5)
    model= GaussianNB()
    score=[]
    for train_index, test_index in kf.split(X,Y):
        X_train, X_test= X.iloc[train_index], X.iloc[test_index]
        Y_train, Y_test= Y.iloc[train_index], Y.iloc[test_index]
        model.fit(X_train, Y_train)
        Y_pred= model.predict(X_test)
        score.append(accuracy_score(Y_test, Y_pred))
    return score    
        

score_cv_model4= k_folds(X_train, Y_train)
score_cv_model4=np.mean(score_cv_model4)
model4= GaussianNB()
model4.fit(X_train, Y_train)
pred4= model4.predict(X_test)
print(accuracy_score(Y_test, pred4))
print(classification_report(Y_test,pred4))
```

```python
float(cv1.best_score_), float(cv2.best_score_), float(cv3.best_score_), float(score_cv_model4)
```
